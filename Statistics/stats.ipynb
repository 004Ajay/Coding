# Histograms

Consider we measured height of students in a class and plotted on a single line, naturally some points will be on top of other or partially hiding other points. 

We can solve this by plotting it in 2D but the points won't come in a straight line up from a value on the x axis due to the artial hiding problem.

To solve this, we can bin the points, for example, 

Let bins be 100-130, 130-150 and 150-170, so points 123 in 100-130, 164 in  150-170 etc...and points at edge with decimal values (ex, 130.7) can be floor-ed to include in respective bin.

```python
heights = [random.randint(110, 200) for _ in range(500)] # 500 random heights between 110 and 200 cm

sns.histplot(heights, bins=10, kde=True)
plt.title('Distribution of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
mean = 0
std_dev = 1
size = 1000

data = np.random.normal(mean, std_dev, size)

plt.hist(data, bins=50, histtype=, color='blue')
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
